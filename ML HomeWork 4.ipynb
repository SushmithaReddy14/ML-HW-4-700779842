```python
# for Question 1

import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Runtime input
text = input("Enter text: ")

# Process text
doc = nlp(text)

# Stopwords
stopwords = nlp.Defaults.stop_words

# Tokenize → remove stopwords → lemmatize → keep nouns & verbs
result = [
    token.lemma_
    for token in doc
    if token.text.lower() not in stopwords
    and token.is_alpha
    and token.pos_ in ["NOUN", "VERB"]
]

print("Final Output:", result)
```

```python
# for question 2

import spacy

# Load NLP model
nlp = spacy.load("en_core_web_sm")

# Runtime input
text = input("Enter text: ")

# Process text
doc = nlp(text)

# -------------------------
# 1. Named Entity Recognition
# -------------------------
print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)

# -------------------------
# 2. Pronoun ambiguity check
# -------------------------
pronouns = {"he", "she", "they", "him", "her", "them"}

if any(token.text.lower() in pronouns for token in doc):
    print("\nWarning: Possible pronoun ambiguity detected!")
```
